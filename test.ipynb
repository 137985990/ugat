```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 
years = list(range(2016, 2025))
balances = [23707.00, 18436.00, 16401.00, 17462.00, 20181.00, 22913.00, 21528.20, 22953.23, 2015.96]

# yx
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6), gridspec_kw={'height_ratios': [3, 1]})

# 
ax1.plot(years, balances, marker='o', color='blue')
ax1.set_ylim(15000, 25000)
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(labeltop=True)
ax1.xaxis.tick_top()

# 
for x, y in zip(years[:-1], balances[:-1]):  # 2024
    ax1.text(x, y + 300, f'${y:,.0f}', ha='center', va='bottom', fontsize=9)

# 
ax2.plot(years, balances, marker='o', color='blue')
ax2.set_ylim(0, 4000)
ax2.spines['top'].set_visible(False)
ax2.xaxis.tick_bottom()

# 
x, y = years[-1], balances[-1]
ax2.text(x, y + 200, f'${y:,.0f}', ha='center', va='bottom', fontsize=9)

# 
d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

# 
fig.suptitle("Account Balance as of Sep 30 (with Break in Y-axis for 2024)", fontsize=14)
ax2.set_xlabel("Year")
ax1.set_ylabel("Balance ($)")
ax2.set_ylabel("Balance ($)")
ax2.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout()
plt.subplots_adjust(hspace=0.05)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# 
years = list(range(2016, 2025))
balances = [23707.00, 18436.00, 16401.00, 17462.00, 20181.00, 22913.00, 21528.20, 22953.23, 2015.96]
balance_labels = [f'${b:,.2f}' for b in balances]

# 
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 7), gridspec_kw={'height_ratios': [3, 1]})

# 
ax1.plot(years, balances, marker='o', color='blue')
ax1.set_ylim(15000, 25000)
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(labeltop=True)
ax1.xaxis.tick_top()

# 
for x, y in zip(years[:-1], balances[:-1]):
    ax1.text(x, y + 300, f'${y:,.0f}', ha='center', va='bottom', fontsize=9)

# 
ax2.plot(years, balances, marker='o', color='blue')
ax2.set_ylim(0, 4000)
ax2.spines['top'].set_visible(False)
ax2.xaxis.tick_bottom()

# 
x, y = years[-1], balances[-1]
ax2.text(x, y + 200, f'${y:,.0f}', ha='center', va='bottom', fontsize=9)

# 
d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

# 

ax1.set_ylabel("Balance ($)")
ax2.set_ylabel("Balance ($)")
ax2.set_xlabel("Year")
ax2.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# 
table_data = [balance_labels]
table = plt.table(cellText=table_data,
                  colLabels=years,
                  cellLoc='center',
                  loc='bottom',
                  bbox=[0.05, -1, 0.9, 0.5])  # [left, bottom, width, height]

table.auto_set_font_size(False)
table.set_fontsize(9)

plt.tight_layout()
plt.subplots_adjust(hspace=0.05, bottom=0.3)  # 
plt.show()
```
